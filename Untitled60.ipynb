```python
# Complete Colab-ready implementation (code only)
# Advanced Time Series Forecasting: State-Space + Kalman Filter + MLE + SARIMA baseline
# Paste into Colab. Uncomment installs in Colab if needed.

# --- Uncomment in Colab if packages missing ---
# !pip install -q pmdarima statsmodels matplotlib numpy pandas scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy import stats
import os
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)

# Optional SARIMA baseline
try:
    import pmdarima as pm
except Exception:
    pm = None

# -----------------------------
# 0) Configuration & true params
# -----------------------------
n = 600                    # >= 500
season_period = 50
t = np.arange(n)
h = 20                     # forecast horizon for evaluation

# True parameters (documented via print)
true_level0 = 1.5
true_slope0 = 0.003
true_beta_sin = 4.0
true_beta_cos = -2.0
rho_true = 0.995

sigma_w_level = 0.002
sigma_w_slope = 0.0005
sigma_w_season = 0.004
sigma_v = 0.08

print("TRUE PARAMETERS:")
print(f" level0={true_level0}, slope0={true_slope0}")
print(f" beta_sin={true_beta_sin}, beta_cos={true_beta_cos}, season_period={season_period}")
print(f" rho={rho_true}")
print(f" sigma_w_level={sigma_w_level}, sigma_w_slope={sigma_w_slope}, sigma_w_season={sigma_w_season}, sigma_v={sigma_v}")
print("-" * 60)

# -----------------------------
# 1) Synthetic data generation
# -----------------------------
omega = 2 * np.pi / season_period
season_sin = np.sin(omega * t)
season_cos = np.cos(omega * t)

level = np.zeros(n)
slope = np.zeros(n)
beta_sin = np.zeros(n)
beta_cos = np.zeros(n)
y = np.zeros(n)

level[0] = true_level0
slope[0] = true_slope0
beta_sin[0] = true_beta_sin
beta_cos[0] = true_beta_cos
y[0] = level[0] + beta_sin[0]*season_sin[0] + beta_cos[0]*season_cos[0] + np.random.normal(0, sigma_v)

for i in range(1, n):
    level[i] = level[i-1] + slope[i-1] + np.random.normal(0, sigma_w_level)
    slope[i] = slope[i-1] + np.random.normal(0, sigma_w_slope)
    beta_sin[i] = rho_true * beta_sin[i-1] + np.random.normal(0, sigma_w_season)
    beta_cos[i] = rho_true * beta_cos[i-1] + np.random.normal(0, sigma_w_season)
    y[i] = level[i] + beta_sin[i]*season_sin[i] + beta_cos[i]*season_cos[i] + np.random.normal(0, sigma_v)

df = pd.DataFrame({
    "t": t, "y": y,
    "level_true": level, "slope_true": slope,
    "beta_sin_true": beta_sin, "beta_cos_true": beta_cos,
    "season_sin": season_sin, "season_cos": season_cos
})

# Ensure ts variable exists (fixes 'ts not defined')
ts = df['y']

# -----------------------------
# 2) Kalman filter from scratch
# State vector: x = [level, slope, beta_sin, beta_cos]
# -----------------------------
def kalman_filter(y, season_sin, season_cos, params, init_scale=100.0):
    T = len(y)
    rho = params['rho']
    q_level = params['q_level']
    q_slope = params['q_slope']
    q_season = params['q_season']
    r = params['r']

    F = np.array([[1., 1., 0., 0.],
                  [0., 1., 0., 0.],
                  [0., 0., rho, 0.],
                  [0., 0., 0., rho]])
    Q = np.diag([q_level**2, q_slope**2, q_season**2, q_season**2])
    R = r**2

    x_pred = np.zeros((T, 4))
    P_pred = np.zeros((T, 4, 4))
    x_filt = np.zeros((T, 4))
    P_filt = np.zeros((T, 4, 4))

    # Initialization
    x_pred[0] = np.array([np.mean(y[:min(20,len(y))]), 0., 0., 0.])
    P_pred[0] = np.eye(4) * init_scale

    loglik = 0.0
    for tidx in range(T):
        H_t = np.array([1., 0., season_sin[tidx], season_cos[tidx]]).reshape(1,4)
        if tidx > 0:
            x_pred[tidx] = F.dot(x_filt[tidx-1])
            P_pred[tidx] = F.dot(P_filt[tidx-1]).dot(F.T) + Q

        y_pred = H_t.dot(x_pred[tidx])[0]
        S = H_t.dot(P_pred[tidx]).dot(H_t.T)[0,0] + R
        if S <= 0:
            S = 1e-8
        v = y[tidx] - y_pred
        K = P_pred[tidx].dot(H_t.T) / S
        x_filt[tidx] = x_pred[tidx] + (K.flatten() * v)
        P_filt[tidx] = P_pred[tidx] - K.dot(H_t).dot(P_pred[tidx])
        loglik += -0.5 * (np.log(2*np.pi*S) + (v**2)/S)

    return {'x_pred': x_pred, 'P_pred': P_pred, 'x_filt': x_filt, 'P_filt': P_filt, 'loglik': loglik}

# -----------------------------
# 3) Kalman smoother (RTS)
# -----------------------------
def kalman_smoother(kf_out, params):
    x_pred = kf_out['x_pred']
    P_pred = kf_out['P_pred']
    x_filt = kf_out['x_filt']
    P_filt = kf_out['P_filt']
    T = x_filt.shape[0]
    rho = params['rho']
    F = np.array([[1., 1., 0., 0.],
                  [0., 1., 0., 0.],
                  [0., 0., rho, 0.],
                  [0., 0., 0., rho]])
    x_smooth = np.zeros_like(x_filt)
    P_smooth = np.zeros_like(P_filt)
    x_smooth[-1] = x_filt[-1]
    P_smooth[-1] = P_filt[-1]
    for k in range(T-2, -1, -1):
        P_pred_kp1 = P_pred[k+1]
        # Smoother gain
        Ck = P_filt[k].dot(F.T).dot(np.linalg.pinv(P_pred_kp1))
        x_smooth[k] = x_filt[k] + Ck.dot(x_smooth[k+1] - x_pred[k+1])
        P_smooth[k] = P_filt[k] + Ck.dot(P_smooth[k+1] - P_pred_kp1).dot(Ck.T)
    return {'x_smooth': x_smooth, 'P_smooth': P_smooth}

# -----------------------------
# 4) Negative log-likelihood (optimize over transformed params)
# theta = [log q_level, log q_slope, log q_season, log r, atanh(rho/0.999)]
# -----------------------------
def neg_loglik_from_params_vector(theta, y, season_sin, season_cos):
    q_level = np.exp(theta[0])
    q_slope = np.exp(theta[1])
    q_season = np.exp(theta[2])
    r = np.exp(theta[3])
    rho = np.tanh(theta[4]) * 0.999
    params = {'rho': rho, 'q_level': q_level, 'q_slope': q_slope, 'q_season': q_season, 'r': r}
    out = kalman_filter(y, season_sin, season_cos, params)
    nll = - out['loglik']
    penalty = 1e-8 * (q_level + q_slope + q_season + r)
    return nll + penalty

# -----------------------------
# 5) Sanity: loglik with TRUE params
# -----------------------------
true_params = {'rho': rho_true, 'q_level': sigma_w_level, 'q_slope': sigma_w_slope, 'q_season': sigma_w_season, 'r': sigma_v}
kf_true = kalman_filter(df['y'].values, df['season_sin'].values, df['season_cos'].values, true_params)
print(f"Log-likelihood with TRUE params: {kf_true['loglik']:.6f}")

# -----------------------------
# 6) Train/test split and MLE fit on train
# -----------------------------
train_size = int(n * 0.8)
y_train = df['y'].values[:train_size]
y_test = df['y'].values[train_size:train_size+h]
season_sin_train = df['season_sin'].values[:train_size]
season_cos_train = df['season_cos'].values[:train_size]

# init close to true but not exact
init_theta = np.log([0.01, 0.001, 0.005, 0.1])
init_theta = np.concatenate([init_theta, [np.arctanh(0.98/0.999)]])

res = minimize(neg_loglik_from_params_vector, init_theta,
               args=(y_train, season_sin_train, season_cos_train),
               method='L-BFGS-B', options={'maxiter':800, 'ftol':1e-9})

if not res.success:
    print("Warning: optimization did not converge:", res.message)

opt_theta = res.x
q_level_mle = np.exp(opt_theta[0])
q_slope_mle = np.exp(opt_theta[1])
q_season_mle = np.exp(opt_theta[2])
r_mle = np.exp(opt_theta[3])
rho_mle = np.tanh(opt_theta[4]) * 0.999

mle_params_train = {'rho': rho_mle, 'q_level': q_level_mle, 'q_slope': q_slope_mle, 'q_season': q_season_mle, 'r': r_mle}
print("\nMLE (train) estimated parameters:")
for k,v in mle_params_train.items():
    print(f" {k}: {v:.6g}")

# -----------------------------
# 7) Get filtered state at end of train and multistep forecast
# -----------------------------
kf_train = kalman_filter(y_train, season_sin_train, season_cos_train, mle_params_train)
x_last = kf_train['x_filt'][-1]

F_mle = np.array([[1., 1., 0., 0.],
                  [0., 1., 0., 0.],
                  [0., 0., rho_mle, 0.],
                  [0., 0., 0., rho_mle]])

def multistep_forecast(x_start, F, season_sin_vals, season_cos_vals, steps):
    x = x_start.copy()
    forecasts = []
    for k in range(steps):
        x = F.dot(x)
        H = np.array([1., 0., season_sin_vals[k], season_cos_vals[k]])
        y_hat = H.dot(x)
        forecasts.append(y_hat)
    return np.array(forecasts)

season_sin_fore = df['season_sin'].values[train_size:train_size+h]
season_cos_fore = df['season_cos'].values[train_size:train_size+h]
y_fore_mle = multistep_forecast(x_last, F_mle, season_sin_fore, season_cos_fore, h)

# True-params forecast (for controlled comparison)
kf_train_true = kalman_filter(y_train, season_sin_train, season_cos_train, true_params)
x_last_true = kf_train_true['x_filt'][-1]
F_true = np.array([[1.,1.,0.,0.],[0.,1.,0.,0.],[0.,0.,rho_true,0.],[0.,0.,0.,rho_true]])
y_fore_true = multistep_forecast(x_last_true, F_true, season_sin_fore, season_cos_fore, h)

# -----------------------------
# 8) SARIMA baseline via auto_arima (optional)
# -----------------------------
y_fore_sarima = None
if pm is not None:
    try:
        sarima = pm.auto_arima(pd.Series(y_train), seasonal=True, m=season_period,
                               stepwise=True, suppress_warnings=True, error_action='ignore', max_order=12)
        y_fore_sarima = sarima.predict(n_periods=h)
    except Exception as e:
        print("Warning: auto_arima failed:", str(e))

# -----------------------------
# 9) Metrics
# -----------------------------
def rmse(a,b): return np.sqrt(np.mean((a-b)**2))
def mae(a,b): return np.mean(np.abs(a-b))
def mape(a,b): return np.mean(np.abs((a-b) / (np.where(np.abs(b)<1e-8,1e-8,b)))) * 100.0

metrics = []
metrics.append({'method':'Kalman (MLE train)', 'RMSE':rmse(y_fore_mle, y_test), 'MAE':mae(y_fore_mle, y_test), 'MAPE_pct':mape(y_fore_mle, y_test)})
metrics.append({'method':'Kalman (True params)', 'RMSE':rmse(y_fore_true, y_test), 'MAE':mae(y_fore_true, y_test), 'MAPE_pct':mape(y_fore_true, y_test)})
if y_fore_sarima is not None:
    metrics.append({'method':'SARIMA', 'RMSE':rmse(y_fore_sarima, y_test), 'MAE':mae(y_fore_sarima, y_test), 'MAPE_pct':mape(y_fore_sarima, y_test)})

metrics_df = pd.DataFrame(metrics)

param_table = pd.DataFrame([
    {'param':'rho', 'true':true_params['rho'], 'mle_train':mle_params_train['rho']},
    {'param':'q_level (std)', 'true':true_params['q_level'], 'mle_train':mle_params_train['q_level']},
    {'param':'q_slope (std)', 'true':true_params['q_slope'], 'mle_train':mle_params_train['q_slope']},
    {'param':'q_season (std)', 'true':true_params['q_season'], 'mle_train':mle_params_train['q_season']},
    {'param':'r (obs std)', 'true':true_params['r'], 'mle_train':mle_params_train['r']},
])

# Save outputs
os.makedirs('/mnt/data', exist_ok=True)
df.to_csv('/mnt/data/synthetic_timeseries.csv', index=False)
metrics_df.to_csv('/mnt/data/timeseries_metrics.csv', index=False)
param_table.to_csv('/mnt/data/params_comparison.csv', index=False)

# -----------------------------
# 10) Print quick summaries
# -----------------------------
print("\nForecast (test) - truth vs mle_fore vs true_fore (first {} steps):".format(h))
for i in range(h):
    s = f"t={train_size+i}: truth={y_test[i]:.6f}, mle={y_fore_mle[i]:.6f}, true={y_fore_true[i]:.6f}"
    if y_fore_sarima is not None:
        s += f", sarima={y_fore_sarima[i]:.6f}"
    print(s)

print("\nMetrics:")
print(metrics_df.to_string(index=False))
print("\nParameter comparison (true vs mle_train):")
print(param_table.to_string(index=False))
print("\nSaved CSVs to /mnt/data: synthetic_timeseries.csv, timeseries_metrics.csv, params_comparison.csv")

# -----------------------------
# 11) Residual diagnostics (standardized innovations on train)
# -----------------------------
kf_train_full = kalman_filter(y_train, season_sin_train, season_cos_train, mle_params_train)
innovations = []
S_list = []
for tidx in range(len(y_train)):
    H_t = np.array([1., 0., season_sin_train[tidx], season_cos_train[tidx]]).reshape(1,4)
    xpred = kf_train_full['x_pred'][tidx]
    Ppred = kf_train_full['P_pred'][tidx]
    ypred = H_t.dot(xpred)[0]
    S = float(H_t.dot(Ppred).dot(H_t.T)[0,0] + mle_params_train['r']**2)
    innovations.append(y_train[tidx] - ypred)
    S_list.append(S)
innovations = np.array(innovations)
std_innov = np.sqrt(np.array(S_list))
std_innov_adj = innovations / std_innov

# Save standardized innovations
pd.DataFrame({'std_innov': std_innov_adj}).to_csv('/mnt/data/standardized_innovations.csv', index=False)

# -----------------------------
# 12) Minimal plots (optional in Colab)
# -----------------------------
try:
    import matplotlib
    plt.figure(figsize=(12,4))
    plt.plot(df['t'], df['y'], label='Observations', linewidth=1)
    plt.plot(df['t'][:train_size], df['y'][:train_size], label='Train', linewidth=2)
    plt.plot(df['t'][train_size:train_size+h], y_fore_mle, label='Forecast (MLE)', marker='o')
    plt.plot(df['t'][train_size:train_size+h], y_fore_true, label='Forecast (True)', marker='x')
    if y_fore_sarima is not None:
        plt.plot(df['t'][train_size:train_size+h], y_fore_sarima, label='SARIMA', marker='d')
    plt.axvline(train_size, color='k', linestyle='--', alpha=0.6)
    plt.legend(); plt.title('Observations & multi-step forecasts'); plt.show()
except Exception:
    pass

# End of script
```
